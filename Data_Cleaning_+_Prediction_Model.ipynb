# Predict the speed

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Load data into a pandas dataframe
data = pd.read_csv('/content/drive/MyDrive/Travel/travel-times.csv')

# Assign the target variable (y) and predictor variable (X)
y = data['MaxSpeed']
X = data[['Distance', 'AvgSpeed']]

# Create a random forest regression model and fit it to the data
model = RandomForestRegressor().fit(X, y)

# Make predictions on the training data
y_pred = model.predict(X)

# Predict the speed using the fitted model
new_data = [[10, 500], [20, 1000], [30, 1500]] # example data for prediction
predictions = model.predict(new_data)
print(predictions)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load data
data = pd.read_csv('/content/drive/MyDrive/Travel/travel-times.csv')

# Select variables X and Y
X = data[['Distance', 'AvgSpeed']]
Y = data['MaxSpeed'].values.reshape(-1, 1)

# Build linear regression model
model = LinearRegression().fit(X, Y)

# Make predictions on the entire dataset
Y_pred = model.predict(X)

# Compute mean squared error
mse = mean_squared_error(Y, Y_pred)

print('Mean squared error:', mse)
```

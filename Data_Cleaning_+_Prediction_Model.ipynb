```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load data
data = pd.read_csv('/content/drive/MyDrive/Travel/travel-times.csv')

# Select variables X and Y
X = data[['Distance', 'MaxSpeed']]
Y = data['AvgMovingSpeed'].values.reshape(-1, 1)

# Build linear regression model
model = LinearRegression().fit(X, Y)

# Predict speed for a distance of 70 km and max speed of 100 km/h
distance = 70
max_speed = 100
predicted_speed = model.predict([[distance, max_speed]])

print('Predicted speed:', predicted_speed[0][0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load data
data = pd.read_csv('/content/drive/MyDrive/Travel/travel-times.csv')

# Select variables X and Y
X = data[['Distance', 'MaxSpeed']]
Y = data['AvgMovingSpeed'].values.reshape(-1, 1)

# Build linear regression model
model = LinearRegression().fit(X, Y)

# Make predictions on the entire dataset
Y_pred = model.predict(X)

# Compute mean squared error
mse = mean_squared_error(Y, Y_pred)

print('Mean squared error:', mse)
```

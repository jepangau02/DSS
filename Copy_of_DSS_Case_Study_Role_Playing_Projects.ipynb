```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Data
x = np.array([46.8, 52.1, 55.1, 59.2, 61.9, 66.2, 69.9, 76.8, 79.7, 79.3, 80.2, 83.3]).reshape((-1, 1))
y = np.array([12530, 10800, 10180, 9730, 9750, 10230, 11160, 13910, 15110, 15690, 17020, 17880])

# Fit linear regression model
model = LinearRegression().fit(x, y)

# Get parameters of the regression line
a = model.intercept_
b = model.coef_[0]

# Calculate R-squared value
r_sq = model.score(x, y)

# Calculate mean squared error and root mean squared error
y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

# Plot data and regression line
plt.scatter(x, y)
plt.plot(x, a + b*x, color='red')
plt.xlabel("Outside high temperature (X)")
plt.ylabel("Daily electricity usage (kilowatts)")
plt.show()

# Print parameters of the regression line, R-squared value, and error metrics
print("Regression line equation: Y = {:.2f} + {:.2f}X".format(a, b))
print("R-squared value: {:.4f}".format(r_sq))
print("Mean squared error: {:.2f}".format(mse))
print("Root mean squared error: {:.2f}".format(rmse))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data
x = np.array([46.8, 52.1, 55.1, 59.2, 61.9, 66.2, 69.9, 76.8, 79.7, 79.3, 80.2, 83.3]).reshape((-1, 1))
y = np.array([12530, 10800, 10180, 9730, 9750, 10230, 11160, 13910, 15110, 15690, 17020, 17880])

# Create polynomial features for quadratic regression
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# Fit quadratic regression model
model = LinearRegression().fit(x_poly, y)

# Get parameters of the quadratic regression curve
a = model.intercept_
b1, b2 = model.coef_[1:]

# Plot data and quadratic regression curve
plt.scatter(x, y)
plt.plot(x, a + b1*x + b2*x**2, color='red')
plt.xlabel("Outside high temperature (X)")
plt.ylabel("Daily electricity usage (kilowatts)")
plt.show()

# Print parameters of the quadratic regression curve
print("Quadratic regression curve equation: Y = {:.2f} + {:.2f}X + {:.2f}X^2".format(a, b1, b2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

# Data
x = np.array([46.8, 52.1, 55.1, 59.2, 61.9, 66.2, 69.9, 76.8, 79.7, 79.3, 80.2, 83.3]).reshape((-1, 1))
y = np.array([12530, 10800, 10180, 9730, 9750, 10230, 11160, 13910, 15110, 15690, 17020, 17880])

# Create polynomial features for quadratic regression
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# Fit quadratic regression model
model = LinearRegression().fit(x_poly, y)

# Get parameters of the quadratic regression curve
a = model.intercept_
b1, b2 = model.coef_[1:]

# Calculate R-squared value, mean squared error, and root mean squared error
y_pred = model.predict(x_poly)
r_sq = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

# Plot data and quadratic regression curve
plt.scatter(x, y)
plt.plot(x, a + b1*x + b2*x**2, color='red')
plt.xlabel("Outside high temperature (X)")
plt.ylabel("Daily electricity usage (kilowatts)")
plt.show()

# Print parameters of the quadratic regression curve, R-squared value, and error metrics
print("Quadratic regression curve equation: Y = {:.2f} + {:.2f}X + {:.2f}X^2".format(a, b1, b2))
print("R-squared value: {:.4f}".format(r_sq))
print("Mean squared error: {:.2f}".format(mse))
print("Root mean squared error: {:.2f}".format(rmse))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define the data for section 1
X1 = np.array([46.8, 52.1, 55.1])
Y1 = np.array([12530, 10800, 10180])

# Fit a linear regression model
model1 = LinearRegression().fit(X1.reshape(-1, 1), Y1)

# Print the coefficients
print('Section 1: slope =', model1.coef_[0], ', intercept =', model1.intercept_)

# Calculate R-squared
y_pred1 = model1.predict(X1.reshape(-1, 1))
r_squared1 = 1 - np.sum((Y1 - y_pred1)**2) / np.sum((Y1 - np.mean(Y1))**2)
print('Section 1: R-squared =', r_squared1)

# Calculate RMSE
mse1 = np.mean((Y1 - y_pred1)**2)
rmse1 = np.sqrt(mse1)
print('Section 1: RMSE =', rmse1)

plt.scatter(X1, Y1)
plt.plot(X1, y_pred1, color='red')
plt.title('Section 1')
plt.xlabel('Temperature')
plt.ylabel('Kilowatts')
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the data for section 3
X3 = np.array([76.8, 79.7, 79.3, 80.2, 83.3])
Y3 = np.array([13910, 15110, 15690, 17020, 17880])

# Define the quadratic function
def quadratic_function(x, a, b, c):
    return a*x**2 + b*x + c

# Fit the quadratic function
popt, pcov = curve_fit(quadratic_function, X3, Y3)

# Print the coefficients
print('Section 3: a =', popt[0], ', b =', popt[1], ', c =', popt[2])

# Calculate R-squared
y_pred3 = quadratic_function(X3, *popt)
r_squared3 = 1 - np.sum((Y3 - y_pred3)**2) / np.sum((Y3 - np.mean(Y3))**2)
print('Section 3: R-squared =', r_squared3)

# Calculate RMSE
mse3 = np.mean((Y3 - y_pred3)**2)
rmse3 = np.sqrt(mse3)
print('Section 3: RMSE =', rmse3)

# Plot the data and the quadratic function
plt.scatter(X3, Y3, label='data')
plt.plot(X3, quadratic_function(X3, *popt), 'r', label='quadratic fit')
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Load the data
X = np.array([46.8, 52.1, 55.1, 59.2, 61.9, 66.2, 69.9, 76.8, 79.7, 79.3, 80.2, 83.3])
Y = np.array([12530, 10800, 10180, 9730, 9750, 10230, 11160, 13910, 15110, 15690, 17020, 17880])

# Scale the data
X_scaled = (X - X.min()) / (X.max() - X.min())
Y_scaled = (Y - Y.min()) / (Y.max() - Y.min())

# Build the neural network
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=[1], activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1)
])

# Compile the model
model.compile(loss='mse', optimizer='adam')
model.save('DSS.h5')

# Train the model
model.fit(X_scaled, Y_scaled, epochs=1500, verbose=0)

# Make predictions
Y_pred_scaled = model.predict(X_scaled)
Y_pred = Y_pred_scaled * (Y.max() - Y.min()) + Y.min()

# Print the predictions
print(Y_pred)

plt.scatter(X, Y, color='green')
plt.scatter(X, Y_pred, color='red')
plt.title('Using Neural Network')
plt.xlabel('Temperature')
plt.ylabel('Kilowatts')
plt.show()
```
